# What is Beta Distribution?

The Beta distribution is a continuous probability distribution defined on the interval [0, 1]. It is commonly used to model random variables that represent probabilities or proportions. The distribution is characterized by two shape parameters, α (alpha) and β (beta), which control the shape of the distribution.

Key features of the Beta distribution:
1. Versatile shape: Can take various shapes depending on the values of α and β.
2. Bounded range: Always between 0 and 1, making it suitable for modeling probabilities.
3. Conjugate prior: Often used in Bayesian inference as a conjugate prior for binomial and Bernoulli distributions.
4. Relationship to other distributions: Special cases include the uniform distribution and arcsine distribution.

The Beta distribution has applications in various fields, including:
- Bayesian statistics
- Project management (modeling task completion times)
- Reliability analysis
- Marketing (modeling customer behavior)
- Finance (modeling asset returns)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Set up the parameters
alphas = [0.5, 1, 2, 5]
betas = [0.5, 1, 2, 5]
x = np.linspace(0, 1, 1000)

# Create the plot
plt.figure(figsize=(12, 8))

# Plot different Beta distributions
for a, b in zip(alphas, betas):
    y = beta.pdf(x, a, b)
    plt.plot(x, y, label=f'α={a}, β={b}')

plt.title('Beta Distribution - Various Parameters')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Example: Generate random samples
np.random.seed(42)
sample_size = 1000
random_samples = beta.rvs(2, 5, size=sample_size)

plt.figure(figsize=(10, 6))
plt.hist(random_samples, bins=30, density=True, alpha=0.7)
plt.title('Histogram of Beta(2, 5) Random Samples')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

print(f"Mean of samples: {np.mean(random_samples):.4f}")
print(f"Variance of samples: {np.var(random_samples):.4f}")
```
