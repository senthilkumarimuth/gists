# Gamma Distribution

The Gamma distribution is a continuous probability distribution that is commonly used to model positive-valued random variables. It is particularly useful in various fields, including finance, engineering, and natural sciences.

Key characteristics of the Gamma distribution:

1. Shape parameter (k or α): Controls the shape of the distribution.
2. Scale parameter (θ) or Rate parameter (β = 1/θ): Determines the spread of the distribution.

The probability density function (PDF) of the Gamma distribution is given by:

f(x; k, θ) = (x^(k-1) * e^(-x/θ)) / (θ^k * Γ(k))

Where:
- x > 0 is the random variable
- k > 0 is the shape parameter
- θ > 0 is the scale parameter
- Γ(k) is the Gamma function

Some important properties of the Gamma distribution:

1. Mean: k * θ
2. Variance: k * θ^2
3. Skewness: 2 / sqrt(k)
4. Kurtosis: 6 / k

The Gamma distribution is versatile and includes several special cases:
- When k = 1, it becomes the Exponential distribution
- When k is an integer, it represents the Erlang distribution
- The Chi-squared distribution is a special case of the Gamma distribution

In the following cells, we'll explore the Gamma distribution using Python and visualize its properties.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Define the shape and scale parameters
k_values = [1, 2, 5, 10]  # Shape parameter
theta = 2  # Scale parameter

# Create a range of x values
x = np.linspace(0, 20, 1000)

# Plot the Gamma distribution for different shape parameters
plt.figure(figsize=(10, 6))

for k in k_values:
    y = gamma.pdf(x, a=k, scale=theta)
    plt.plot(x, y, label=f'k={k}, θ={theta}')

plt.title('Gamma Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Calculate and print some properties of the Gamma distribution
for k in k_values:
    mean = k * theta
    variance = k * theta**2
    skewness = 2 / np.sqrt(k)
    kurtosis = 6 / k
    
    print(f"\nFor k={k} and θ={theta}:")
    print(f"Mean: {mean:.2f}")
    print(f"Variance: {variance:.2f}")
    print(f"Skewness: {skewness:.2f}")
    print(f"Kurtosis: {kurtosis:.2f}")
```
