# Chi-Square Distribution

The Chi-Square distribution is a continuous probability distribution that is commonly used in statistical hypothesis testing and inferential statistics. It is derived from the sum of squares of independent standard normal random variables.

Key characteristics of the Chi-Square distribution:

1. It is always non-negative and right-skewed.
2. It has one parameter called degrees of freedom (df).
3. As the degrees of freedom increase, the distribution becomes more symmetric and approaches a normal distribution.

Common applications of the Chi-Square distribution include:

- Goodness-of-fit tests
- Tests of independence in contingency tables
- Estimating variance in a normal distribution
- Confidence interval estimation for population variance

In this notebook, we will explore the properties and applications of the Chi-Square distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set random seed for reproducibility
np.random.seed(42)

# Generate Chi-Square distributed data
df = 5  # degrees of freedom
n_samples = 1000
chi_square_data = np.random.chisquare(df, n_samples)

# Create a histogram of the data
plt.figure(figsize=(10, 6))
plt.hist(chi_square_data, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

# Generate points for the theoretical Chi-Square distribution
x = np.linspace(0, max(chi_square_data), 100)
chi_square_pdf = stats.chi2.pdf(x, df)

# Plot the theoretical Chi-Square distribution
plt.plot(x, chi_square_pdf, 'r-', lw=2, label=f'Chi-Square PDF (df={df})')

plt.title(f'Chi-Square Distribution (df={df})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()

# Calculate some statistics
mean = np.mean(chi_square_data)
variance = np.var(chi_square_data)
skewness = stats.skew(chi_square_data)

print(f"Sample Mean: {mean:.2f}")
print(f"Sample Variance: {variance:.2f}")
print(f"Sample Skewness: {skewness:.2f}")

# Theoretical values
print(f"\nTheoretical Mean: {df}")
print(f"Theoretical Variance: {2*df}")
print(f"Theoretical Skewness: {np.sqrt(8/df):.2f}")
```
