# Tanh Activation Function

The hyperbolic tangent (tanh) activation function is a popular choice in neural networks. It is a smooth, S-shaped function that maps input values to the range (-1, 1). Key characteristics of tanh include:

1. Output range: (-1, 1)
2. Centered at zero
3. Symmetric around the origin
4. Steeper gradient compared to sigmoid function
5. Often used in hidden layers of neural networks

The tanh function is defined as:

tanh(x) = (e^x - e^-x) / (e^x + e^-x)

or

tanh=2 sigmoid(2z)-1


Advantages:
- Zero-centered output
- Handles negative inputs well
- Stronger gradients than sigmoid

Disadvantages:
- Can still suffer from vanishing gradient problem for very large or small inputs
- Computationally more expensive than ReLU

In PyTorch, you can use `torch.nn.Tanh()` to apply the tanh activation function in your neural network models.
