# ReLU Activation Function

The Rectified Linear Unit (ReLU) is a popular activation function in deep learning. It's defined as:

f(x) = max(0, x)

ReLU has several advantages:
1. Simple computation
2. Non-linear behavior
3. Sparse activation
4. Reduces vanishing gradient problem

However, it also has some limitations:
1. "Dying ReLU" problem
2. Not zero-centered

We'll explore the ReLU function and its variants in this notebook.




# Leaky ReLU

Leaky ReLU is a variant of the ReLU activation function designed to address the "dying ReLU" problem. It's defined as:

f(x) = max(αx, x), where α is a small positive constant (typically 0.01)

Key features of Leaky ReLU:
1. Allows small negative values when the input is less than zero
2. Helps prevent neurons from becoming inactive
3. Maintains most of the benefits of standard ReLU
4. Can improve gradient flow in some cases

The main advantage of Leaky ReLU over standard ReLU is that it allows for a small, non-zero gradient when the unit is not active (x < 0), which can help prevent the dying ReLU problem and potentially lead to more robust learning.
