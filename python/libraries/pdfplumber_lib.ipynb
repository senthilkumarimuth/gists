# extract table

```python
import pdfplumber

# Open a PDF file
with pdfplumber.open(r"C:\Users\senthil.marimuthu\OneDrive - HTC Global Services, Inc\self\gists\artificial_intelligence\image_text_parsing\UD1-220224-122107-v2_test.pdf") as pdf:
    # Get the first page of the PDF
    page = pdf.pages[30]
    
    # Extract tables from the page
    tables = page.extract_table()

    # Print the tables
    for row in tables:
        print(row)
```
